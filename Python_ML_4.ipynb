```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/bipulshahi/Dataset/refs/heads/main/dataset.csv',
                header = None)

df.columns = ['x' , 'y']
df.head()
```

```python
plt.figure(figsize=(5,3))
plt.scatter(df['x'] , df['y'])
plt.show()
```

```python
x = df[['x']]
y = df[['y']]

from sklearn.linear_model import LinearRegression
model_a = LinearRegression()
model_a.fit(x,y)

yp = model_a.predict(x)
```

```python
print(model_a.coef_)
print(model_a.intercept_)
```

```python
plt.figure(figsize=(5,3))

plt.scatter(x,y)
plt.plot(x,yp,'r')

plt.show()
```

```python
#Calculate Mean absolute error for model_a

mae_linear_model = abs(y - yp).mean()
print(mae_linear_model)
```

Non - Linear Regression

```python
#Step 1 - Transform your data to fit a non-linear equation
# x => x & x^2

x = df[['x']]
y = df[['y']]
```

```python
x2 = pd.concat((x , x*x , x*x*x) , axis = 1)
x2.columns = ['x' , 'x_squared' , 'x_cube']
x2.head(3)
```

```python
from sklearn.linear_model import LinearRegression
model_b = LinearRegression()
model_b.fit(x2,y)

yp2 = model_b.predict(x2)
```

```python
print(model_b.coef_)
print(model_b.intercept_)
```

```python
plt.figure(figsize=(5,3))

plt.scatter(x,y)
plt.plot(x,yp2,'r')

plt.show()
```

```python
#calculate mean absolute error for model_b

mae_non_linear_model = abs(y - yp2).mean()
print(mae_non_linear_model)
```
